**Phishing Detection Data**

```python
import re

# Keywords commonly used in phishing emails
phishing_keywords = [
    "urgent", "immediate action required", "act now", "limited time",
    "final warning", "account suspended", "account blocked",
    "verify your account", "update your account", "confirm your identity",
    "security alert", "unauthorized login", "suspicious activity detected",
    "password", "login", "enter your credentials", "reset your password",
    "confirm password", "otp", "one time password",
    "bank", "credit card", "debit card", "payment failed",
    "transaction declined", "refund", "invoice attached",
    "click here", "login now", "update now", "verify now",
    "open attachment", "download now",
    "you won", "claim prize", "free gift", "lottery winner",
    "congratulations", "exclusive offer"
]

# Suspicious domains (often used in phishing)
suspicious_domains = [
    "bit.ly", "tinyurl.com", "goo.gl", "t.co", "ow.ly",
    "secure-login", "account-verify", "update-info",
    "free-login", "secure-update", "login-verification",
    "paypal-secure", "banking-update", "verify-account",
    "signin-alert", "user-authentication"
]

# Suspicious attachment extensions
suspicious_attachments = [
    ".exe", ".zip", ".rar", ".scr", ".bat", ".js", ".vbs", ".docm", ".xlsm"
]

# Suspicious email patterns (regex)
suspicious_email_patterns = [
    r"support@.*-secure\.com",
    r"no-reply@.*verify.*",
    r".*@[0-9]+\..*"
]

# -------------------------------
# Detection Function
# -------------------------------

def detect_phishing(subject, body):
    score = 0
    text = (subject + " " + body).lower()

    # Keyword detection
    for word in phishing_keywords:
        if word in text:
            score += 1

    # URL detection
    links = re.findall(r'https?://\S+', body)
    for link in links:
        for domain in suspicious_domains:
            if domain in link:
                score += 2

    # Attachment detection
    for ext in suspicious_attachments:
        if ext in body.lower():
            score += 2

    # Email pattern detection
    for pattern in suspicious_email_patterns:
        if re.search(pattern, subject + body, re.IGNORECASE):
            score += 2

    # -------------------------------
    # Classification
    # -------------------------------
    if score >= 6:
        return "🚨 Phishing Email Detected"
    elif score >= 3:
        return "⚠️ Suspicious Email"
    else:
        return "✅ Safe Email"

# Main AI Agent Loop

def run_agent():
    print("======================================")
    print("  Phishing Email Detective Agent")
    print("  Type 'exit' to quit")
    print("======================================\n")

    while True:
        subject = input("Enter Email Subject: ")
        if subject.lower() == "exit":
            print("Exiting Agent... Stay Safe!")
            break

        body = input("Enter Email Body: ")

        result = detect_phishing(subject, body)

        print("\n--- Analysis Result ---\n")
        print(result)
        print("\n------------------------\n")

# Run Program

if __name__ == "__main__":
    run_agent()
```
