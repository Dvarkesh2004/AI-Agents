**Detect_MaliciousQuery**

```python
import re

# Patterns for common attacks
patterns = {
    "SQL Injection": [
        r"(\bSELECT\b|\bUNION\b|\bINSERT\b|\bDROP\b|\bUPDATE\b|\bDELETE\b)",
        r"(--|#|;)",
        r"(' OR '1'='1)",
        r"(\bOR\b\s+1=1)",
        r"(\bAND\b\s+1=1)",
        r"(\bEXEC\b|\bEXECUTE\b)",
        r"(\bSLEEP\(|\bBENCHMARK\()"
    ],

    "XSS": [
        r"<script.*?>.*?</script>",
        r"javascript:",
        r"onerror=",
        r"onload=",
        r"alert\(",
        r"<img.*?onerror=.*?>",
        r"<iframe.*?>.*?</iframe>"
    ],

    "Command Injection": [
        r"(;|\|\||&&)",
        r"\b(cat|ls|whoami|rm|pwd|id|uname)\b",
        r"`.*?`",
        r"\$\((.*?)\)"
    ],

    "Path Traversal": [
        r"\.\./",
        r"\.\.\\",
        r"/etc/passwd",
        r"C:\\Windows\\System32"
    ],

    "File Inclusion": [
        r"(http|https|ftp)://",
        r"\binclude\b",
        r"\brequire\b",
        r"\.\./.*\.(php|txt|log)"
    ],

    "LDAP Injection": [
        r"(\*|\(|\)|\||&)",
        r"(cn=|uid=|ou=)"
    ],

    "XPath Injection": [
        r"('|\")\s*or\s*('|\")\d=('|\")\d",
        r"//\w+\["
    ],

    "Email Injection": [
        r"\n|\r",
        r"(bcc:|cc:|to:)"
    ],

    "SSRF": [
        r"http://127\.0\.0\.1",
        r"http://localhost",
        r"169\.254\.169\.254",
        r"file://"
    ],

    "Open Redirect": [
        r"(http|https)://.*@",
        r"//.*\."
    ]
}

def detect_attack(user_input):
    for attack_type, regex_list in patterns.items():
        for pattern in regex_list:
            if re.search(pattern, user_input, re.IGNORECASE):
                return f"⚠️ Potential {attack_type} detected!"
    return "✅ Input looks safe."

def run_agent():
    print("Cybersecurity AI Agent (Detection Mode)")
    print("Type 'exit' to stop\n")

    while True:
        user_input = input("Enter query: ")

        if user_input.lower() == "exit":
            print("Exiting...")
            break

        result = detect_attack(user_input)
        print(result)

# Run the agent
if __name__ == "__main__":
    run_agent()
```
